```python
#q5

import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
print(df)
print(df.tail())
print(df.columns)
print(df['Fare'])
print(df[['Age','Fare']])
print(df.describe())
print(df.dtypes)
print(df.dtypes[df.dtypes=="object"].describe())
print(df[['Ticket']][4:10:2])
print(pd.Categorical(df['Age']))
print(df[df['Fare']>10])
print(df[df['Fare']==0])
print(df.iloc[0:89:2])
print(df.loc[0:56,['Sex','Fare']])
print(pd.Categorical(df['Fare']))
```
